```python
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm 
import seaborn as sns
%matplotlib inline

import sys, os, copy
print("Using environment in "+sys.prefix)
print("Python version "+sys.version)
```

```python
# Different dataframe for each city, pull them together in a Dictionary, following Eric's example

filenames={'Phoenix':'Phoenix.xlsx', 'Dallas':'Dallas.xlsx', 'Atlanta':'Atlanta.xlsx', 'Vegas':'Las Vegas.xlsx'}
allCitiesU = {city:pd.DataFrame() for city in filenames.keys()}

for city in filenames.keys():
    allCitiesU[city]=pd.read_excel('Data/Unemployment/'+filenames[city],header=9)
    print(allCitiesU[city].head())


# print(allCities)

```

```python
Employment=pd.DataFrame()                      # A dataframe containing emplyment for all the four metro areas
for city in filenames.keys():
    Employment[city]=allCitiesU[city]['employment']

EmploymentNormalized=copy.deepcopy(Employment)  # This will hold the normalized Emplyment data
EmploymentNormalized=Employment /Employment.iloc[0,:]

print(Employment.head())
print(EmploymentNormalized.head())

#Plots of emplyment and normalized employment, following Eric's example

Employment.plot(colormap=cm.rainbow)

EmploymentNormalized.plot(colormap=cm.rainbow)
```

```python
LaborForce=pd.DataFrame()            # A dataframe containing Labor Force for all the four metro areas
for city in filenames.keys():
    LaborForce[city]=allCitiesU[city]['labor force']
    
LaborForce.head()

sns.pairplot(LaborForce)             # Pairplot of Labor Force for the 4 metro areas

# g = sns.PairGrid(LaborForce)
# g.map(plt.scatter)
```
